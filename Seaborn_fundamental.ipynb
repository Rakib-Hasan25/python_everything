```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px # using for an inbuild dataset
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level function  -> graph look like rectengular shape
sns.scatterplot(data = tips , x='total_bill' , y = 'tip' )
# in mathplot lib we have to write code for leveling but here not need
```

```python
#now we want differentiate between male and female
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex')


# here we can we a legend box appear in the graph but we have to write code for that in mathplotlib
```

```python
# we can also do
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time',size='size')
# size increase dot size  # hue change color of dot # style change design of dot
```

```python
# relplot -> figure level ->  graph look like square shape



# it is always better to use figure level
# because using it we can create many more axes level graph



# legend box(sex,time,etc repesentaion box ) is outside from the graph which doesn't distrub in graph
# but in scatter plot box can over lap with graph


sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time',size='size')
# we give kind = "scatter" for scatter graph
```

```python
tips
```

```python
# plotting univariate histogram

#using axes level

sns.histplot(data=tips, x='total_bill')
```

```python
#using figure level
sns.displot(data=tips, x = 'total_bill', kind='hist')
```

```python
#bins perameter
sns.displot(data=tips, x = 'total_bill', kind='hist', bins=20) # number of bins
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

#i want to know customer came in every day
#typically we can say it  countplot ( categorical data's count)
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')


# histogram for male or female customer the amount of tip they give
```

```python
# to see more clearly

# element -> step
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
```

```python
# now doing a very interestiong thing
# most important data set
titanic = sns.load_dataset('titanic') # default dataset in seaborn
titanic
```

```python
# histogram for different age group
sns.displot(data=titanic, x = 'age', kind='hist')
```

```python
# male and female distribution in age group
sns.displot(data=titanic, x = 'age', kind='hist', element ='step', hue = 'sex')
```

```python
# survived people distributaion in age group
sns.displot(data=titanic, x = 'age', kind='hist', element ='step', hue = 'survived')
```

```python
# faceting using col and row -> not work on histplot function


# tip giving distribution for different gender
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

# KDEplot

###+kernal density estimation

##### kdeplot
#### Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel(function), producing a continuous density estimate---> cover in statistics


```python
sns.kdeplot(data=tips,x='total_bill')

# amount of bills distribution

# it giving probability of entire data set of having bill in a range
# in their , 20 bills having the greater probability
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
#hue
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=5,aspect=2)
```

#RugPlot

#### it is used for suppoting the marginal distribution
####not so much used alone
##### Plot marginal distributions by drawing ticks along the x and y axes.

##### This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill') # set of tik on x axis which represent the density(distribution)
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')

# here we want to have relation between total bils and tip ,
#where bill and tip intersect each other grid show more brighter , this represent
# relationship between total_bill and tip
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')


# if lines are closer then there an greater density
```

#2.Matrix plot


+ HeatMap
+ ClusterMap

```python
gap
```

```python
gap.pivot(index='country', columns='year',values = 'lifeExp')
```

```python
# Heatmap -> # axes level function

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')


plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
#annot

eu_coun= gap[gap['continent']=='Europe'].pivot(index ='country',columns='year',values='lifeExp') #

```

```python
plt.figure(figsize=(15,15))
sns.heatmap(eu_coun, annot=True,linewidth=0.5)#annot give value number
```

```python
#cmap --> changing color theme

plt.figure(figsize=(15,15))
sns.heatmap(eu_coun, annot=True,cmap='summer')
```
