Skiprows/nrows Parameter

```python
import numpy as np
```

```python
# Define the rows to skip
rows_to_skip = [0, 2]  # Adjust as needed



#lambda arguments: expression

# In our case, lambda x: x in rows_to_skip is a lambda function that takes one argument (x)
# and returns True if x is in the rows_to_skip list; otherwise, it returns False.


# Read the CSV file skipping the specified rows
boll = pd.read_csv('/content/bollywood.csv' )
print(boll.head(5))
boll = pd.read_csv('/content/bollywood.csv', skiprows=[1,2] )
#here we can see 0 index represent the heading
print(boll.head(5))

# now if we want to skip 1 to 10 th row
c = np.arange(1,10)
print(c)

boll = pd.read_csv('/content/bollywood.csv', skiprows=c )

print(boll.head(5))







# boll = pd.read_csv('/content/bollywood.csv', skiprows=lambda x: x  )
```
