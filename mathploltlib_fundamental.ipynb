```python
# from a pandas dataframe
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#lables title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F') # hax code
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# giving linewidth and line style

plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#giving line marker
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)# change marker size
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# now if we want to know which line refer which player

# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat') # label will be working if you plt.legend
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper left') #we set the plt.legend location

# plt.legend(loc='best') # where will be the best location for this box
```

```python
# plt.scatter simple function
x = np.linspace(-10,10,50) # from -10 to 10 we have 50 number with equal division

z=10*x + 3
print(z)
y = z + np.random.randint(0,300,50)#from 0 to 300 we have random 50 number

print()
print(y)

```

```python
plt.scatter(x,y)
#from this graph we can see increasing of x value , y value also increase , so they might have linearity relation
```

```python
# plt.scatter on pandas data
df = pd.read_csv('/content/batter.csv')
df = df.head(50) # first 50 data
df
```

#Pie chart
+ Univariate/Bivariate Analysis
+ Categorical vs numerical
+ Use case - To find contibution on a standard scale(100%)

contribution of each category in 100%

```python
# simple data
data = [23,45,100,20,49]


subjects = ['eng','science','maths','sst','hindi'] # use for labeling

plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')


# autopct='%0.1f%% used to see what percentage of run they have made
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
#explode use for moving a portion outside , here it is happen for 1st batsman
plt.show()
```
